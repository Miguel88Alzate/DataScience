### TASK  3: Visualize the relationship between success rate of each orbit type


```python
success_rate_orbit = df[['Orbit','Class']].groupby(['Orbit']).mean()
success_rate_orbit = success_rate_orbit.reset_index()
success_rate_orbit
```

```python
sns.barplot(y='Class', x='Orbit',data=success_rate_orbit)
plt.xlabel("Orbit",fontsize=20)
plt.ylabel("Success rate",fontsize=20)
plt.show()
```

### TASK  6: Visualize the launch success yearly trend


```python
year=[]
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
Extract_year()
df['Date'] = year
df.head()
    
```

```python
year_success_rate = df[['Date','Class']].groupby(['Date']).mean()
year_success_rate = year_success_rate.reset_index()
year_success_rate
```

```python
sns.lineplot(y='Class', x='Date',data=year_success_rate)
plt.xlabel("Date",fontsize=20)
plt.ylabel("Success Rate",fontsize=20)
plt.show()
```

## Features Engineering


```python
#By now, you should obtain some preliminary insights about how each important variable would 
#affect the success rate, we will select the features that will be used in success prediction 
#in the future module.

features = df[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial']]
features.head()
```

### TASK  7: Create dummy variables to categorical columns


```python
features_dummies = pd.get_dummies(features)
features_dummies
```

### TASK  8: Cast all numeric columns to `float64`


```python
features_one_hot = features_dummies.astype(float)
features_one_hot
```

```python
features_one_hot.to_csv('dataset_part_3.csv', index=False)
```
